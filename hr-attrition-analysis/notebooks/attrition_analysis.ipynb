# HR Analytics : Employee Attrition Analysis

# Objective

The goal of this analysis is to understand patterns in that are present in an employee data and identify the possible factors that influence the attrition.

This is an initial exploration to get a sense of the dataset before going deeper into modeling or predictions.

```python
# Core Required Libraries

import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
```

# Data Loading

```python
# Loading Dataset
df = pd.read_csv("/kaggle/input/datasets/pavansubhasht/ibm-hr-analytics-attrition-dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv")

df.head()
```

# Basic Exploration

```python
# Understanding the skelton or the structure of the dataset
df.info()
```

```python
# Summarizing the statistics of numerical columns
df.describe()
```

# Data Quality Check

```python
# Checking for the missing values
df.isnull().sum()
```

# Initial Analysis

```python
# Employees working overtime
overtime_employees = df[df["OverTime"] == "Yes"]

# Average age of overtime employees
avg_age_overtime = overtime_employees["Age"].mean()

print("Average age of employees working overtime:", avg_age_overtime)
```

# Observation

From this small check, it looks like employees who work overtime have a certain age pattern.

This might not directly explain the attrition yet, but it gives a starting point to explore relationships between workload and employee behavior.

# Attrition Distribution Check

```python
# Checking how many employees stayed vs left
attrition_counts = df["Attrition"].value_counts()

print(attrition_counts)
```

```python
# Visualizing how many employees stayed vs left
sns.countplot(x="Attrition", data=df)
plt.title("Attrition Distribution")
plt.show()
```

# Observation

There seems to be a noticeable difference between employees who stayed and those who left.

This imbalance could be important later if we move towards building prediction models.

# Data Cleaning

```python
# Checking the dupicate rows
duplicates = df.duplicated().sum()
print("Duplicated rows:",duplicates)

```

```python
# Dropping the unecessary columns
df = df.drop(columns=['EmployeeCount','Over18','StandardHours'])
```

```python
# Checking the shape after columns
df.shape
```

# Observation

No duplicate rows are found in the dataset.

Some columns such as EmployeeCount, Over18, and StandardHours are removed as they do not contribute meaningful information for analysis.

# Overtime vs Attrition Analysis

```python
# Analyzing the relationship between overtime and attrition

sns.countplot(x='OverTime',hue='Attrition',data=df)
plt.title("Overtime vs Attrition")
plt.show()
```

## Observation

Employees who work overtime show a higher proportion of attrition compared to those who do not.

Although many employees working overtime still stay, the likelihood of leaving appears to be higher in this group, suggesting overtime's workload may influence attrition.

# Salary vs Attrition Analysis

```python
# Analyzing salary distribution with respect to attrition
sns.boxplot(x="Attrition",y="MonthlyIncome",data=df)
plt.title("Salary Distribution by Attrition")
plt.show()
```

## Observation

Employees who left the company tend to have lower median salaries compared to those who stayed.

This indicates that lower compensation may be associated with higher attrition, although the other factors could also influence employee decisions.

# Age Distribution Analysis

```python
# Analyzing the age distribution based on attrition 
sns.histplot(data=df,x='Age',hue='Attrition',kde=True)
plt.title("Age Distribution by Attrition")
plt.show()
```

# Conclusion

From the analysis, several factors appear to be associated with employee attrition.

Employees working overtime tend to show a higher likelihood of leaving, indicating that workload may influence attrition.

Salary also appears to play a role, as employees with lower income levels are more likely to leave compared to those with higher salaries.

Age distribution suggests that younger employees may have a higher tendency towards attrition compared to older age groups.

Overall, attrition seems to be influenced by a combination of **workload**, **compensation**, and **demographic factors** rather than a single variable.
